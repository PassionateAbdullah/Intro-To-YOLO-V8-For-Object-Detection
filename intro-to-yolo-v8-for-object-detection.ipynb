# Testing our model

this next part is optional but recommended

you will need my dataset which you can find [here](https://www.kaggle.com/datasets/philopateergeorgei/cooler) It has a vid which we will need for testing

```python
def model_prediction():
    import cv2
    import torch # You need to import torch to load the model

    VIDEOS_DIR = os.path.join('.', '/kaggle/input/cooler') #this will open our folder

    video_path = os.path.join(VIDEOS_DIR, 'street_-_38590 (360p).mp4') #this is the path of the vid
    video_path_out = '{}_out.mp4'.format(video_path) #this makes mp4 format
    #this next code reads the vid using CV2
    cap = cv2.VideoCapture(video_path)
    ret, frame = cap.read()
    H, W, _ = frame.shape
    out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

    model_path = os.path.join('.', 'runs', 'detect', 'train', 'weights', 'last.pt')

    # You need to load the model using torch.load
    model = YOLO("yolov8n.yaml")
    threshold = 0.5

    # You need to define the gt_box variable
    gt_box = [100, 100, 200, 200] # This is just an example, you can change it as you like

    #this is used to make prediction
    while ret:

        results = model(frame)[0]

        for result in results.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = result

            if score > threshold:
                # Draw the predicted box on the frame
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
                # Draw the class name on the frame
                cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

        # Draw the text 'GT' on the frame
        cv2.putText(frame, 'GT', (gt_box[0], gt_box[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.3, (255, 0, 0), 3, cv2.LINE_AA)

        out.write(frame)
        ret, frame = cap.read()
    cap.release()
    out.release()
```

```python
model_prediction()
```
